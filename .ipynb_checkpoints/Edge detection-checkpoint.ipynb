```python
from skimage import io
import numpy as np
import matplotlib.pyplot as plt
from time import time

# Make matplotlib figures appear inline in the
# notebook rather than in a new window
%matplotlib inline
plt.rcParams['figure.figsize'] = (10.0, 8.0) # set default size of plots
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'gray'

def display(img, size=(5,5)):
    # Show image
    plt.figure(figsize = size)
    plt.imshow(img)
    plt.axis('off')
    plt.show()
    

def multi_display(img_tuple):
    # Dispalys multiple image next to one another
    multi_img = np.concatenate(img_tuple, axis=1)
    num = len(img_tuple)
    
    display(multi_img, size=(5*num, 5))
```

We'll start by reviewing the example from lecture: the iguana and the horizontal derivative.

```python
# Horizontal Sobel filter which approximates the horizontal derivative on the image
h_filter = np.array(
[
    [1, 0, -1],
    [2,0,-2],
    [1, 0, -1]
])

image = cv2.imread('./iguana.png')

# Blur the image with a Gaussian kernel in order to reduce noise
blur_image = cv2.GaussianBlur(image, (5,5), 0)

# Pass through horizontal Sobel filter
result = cv2.filter2D(blur_image, -1, h_filter)

# Brighten the image by doubling image values
display(2 * result)
```

```python
# Vertical Sobel filter which approximates the vertical derivative on the image
v_filter = np.array(
[
    [1, 2, 1],
    [0, 0 ,0],
    [-1, -2, -1]
])

image = cv2.imread('./iguana.png')

# Blur the image with a Gaussian kernel in order to reduce noise
blur_image = cv2.GaussianBlur(image, (5,5), 0)

result = cv2.filter2D(blur_image, -1, v_filter)

# Brighten the image by doubling image values
display(2 * result)
```

```python
def edge_finder(image, direction, threshold = 0, blur = True):
    """
        Args: 
            image: numpy array
            direction: string
            threshold: integer
            blur: boolean
        Returns:
            result: numpy array of same size as image
            
        edge_finder attempts to find strong edges in either the horizontal or 
        vertical direction by optionally bluring the image to reduce noise, applying  
        the appropriate Sobel filter, and then eliminating pixels below a 
        threshold value.
    """
    # Check valididty of direction input
    if direction not in {'h', 'v'}:
        raise TypeError('Direction should be h or v')
     
    # Set filter appropriately 
    filter = h_filter if direction == 'h' else v_filter
    
    # Blur image with Gaussian kernel
    if blur:
        blur_image = cv2.GaussianBlur(image, (5,5), 0)
        result = cv2.filter2D(blur_image, -1, filter)
    else:
        result = cv2.filter2D(image, -1, filter)
    
    # Eliminate weak edges from image by zeroing pixels below threshold
    if threshold > 0:
        # Use numpy's vectorize method to apply function to all pixels
        elim_pixels = np.vectorize(lambda x: 0 if x < threshold else x)
        result = elim_pixels(result)
    
    return result
```

```python
display(2 * edge_finder(image, 'h', threshold=50))
display(2 * edge_finder(image, 'v', threshold=50))
```

```python
prison_bars = cv2.imread('./prison_bars.jpeg')

image_1 = edge_finder(prison_bars, 'h')
image_2 = edge_finder(prison_bars, 'v')

multi_display((image_1, image_2))
```

```python
plt.rcParams['figure.figsize'] = (30.0, 8.0) # set default size of plots
noisy_einstein = cv2.imread('./noisy_einstein.png')

image_1 = edge_finder(noisy_einstein, 'v', blur=False)
image_2 = edge_finder(noisy_einstein, 'v')
image_3 = edge_finder(noisy_einstein, 'v', threshold=20)

multi_display((image_1, image_2, image_3))
```

1. The image is dark, and it's hard to tell where the edges are that we found. How can we improve on this? Can we do this in a single filter?
2. The algorithm finds lots of edges, but we don't care about all of them. Could we keep only the brightest edges somehow? 
3. We used a horizontal derivative filter. Create and run a vertical derivative filter. Should it look similar? Does it? Find an image where the horizontal and vertical derivative filters produce very different output.

1. One way to make the edges more obvious is to simply brighten the image. This doesn't even require using a filter, as we can simply multiply the image values by 2.
2. As demonstrated in `edge_finder`, we can eliminate weaker pixels by removing those pixels below a threshold value. This helps reduce the amount of noise in the final image of edges, but does mean edges are not as connected.
3. While we can still recognize the image as the same iguana, the two different directions for the edge detection produce distinctly different images.
    The prison bar image also demonstrates, very clearly, a difference between the horizontal and vertical edge finding filters.    
    
4. As can be seen, passing a filter without first applying a Gaussian kernel (or some other variety of blurring filter) produces a very noisy edge image that is very hard to use. Applying a kernel, while not perfect, does indeed improve the quality of the final edge image. Using a threshold also helps.
